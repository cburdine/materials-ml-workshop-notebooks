```python
from mp_api.client import MPRester

MP_API_KEY = None # (put your API key instead of None)

YBCO_ID = 'mp-20674'

with MPRester(MP_API_KEY) as mpr:
    structure = \
        mpr.get_structure_by_material_id(YBCO_ID)
```

```python
print(structure)
```

```python
from pymatgen.io.ase import AseAtomsAdaptor as aaa

crystal = aaa.get_atoms(structure)
```

```python
from ase.visualize import view

view(crystal, viewer='x3d', repeat=(3,1,1))
```

```python
with MPRester(MP_API_KEY) as mpr:
    bs = mpr.get_bandstructure_by_material_id(YBCO_ID)
```

```python
from pymatgen.electronic_structure.plotter import BSPlotter

plot = BSPlotter(bs).get_plot()
plot.show()
```

```python
with MPRester(MP_API_KEY) as mpr:
    dos = mpr.get_dos_by_material_id(YBCO_ID)
    
print(dos)
```

```python
from pymatgen.electronic_structure.plotter import DosPlotter

plotter = DosPlotter()
plotter.add_dos('YBCO', dos)
plotter.get_plot(xlim=(-10, 10), ylim=(0,30))
plot.show()
```
